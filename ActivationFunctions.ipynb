Plotting activation functions to aid report

```python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Sets x and y data to be plotted
x = np.linspace(-10, 10, 10000)
y = sigmoid(x)
```

```python
# Plots data with necessary labels/styling
plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Sigmoid output')
plt.grid(linestyle='--', alpha=0.4)
plt.title('Sigmoid Function')

# Displays plot
plt.show()
```
