```python
# Used for plotting data
import matplotlib.pyplot as plt
# Used for formatting y-data
import matplotlib.ticker as ticker
# Used to produce a line of best fit
import numpy as np
```

```python
# Corresponding amount of validation samples for each probability
validationSamples = [1000, 5000, 10000, 20000]
# Calculated probabilities to be plotted
probabilities = [38.800, 24.440, 16.122, 7.9773]
```

```python
# Calculates the coefficients of the line of best fit 
coefficients = np.polyfit(validationSamples, probabilities, 3) # Integer (3) signifies a cubic formula 
# Creates x-axis data for the line of best fit
# Fitted x-axis data is equally spaced between the min/max x-values amongst 10000 different values
xFit = np.linspace(min(validationSamples), max(validationSamples), 10000)
# Produces the y-values for each x using the calculated coefficients
yFit = np.polyval(coefficients , xFit)

# Set figure and axes objects for plot
fig, ax = plt.subplots()

# Plots the calculated probabilities as blue points on the graph
ax.plot(validationSamples, probabilities, 'b.', markersize=8, label="Probability")
# Plots the line of best fit
ax.plot(xFit, yFit, 'r-', label="Line of best fit")
# Plots the legend
ax.legend(["Probability", "Line of best fit"], loc='upper right')


# Plots a label for each axis
plt.ylabel('Result Probability')
plt.xlabel('Validation Samples')
# Plots the title
plt.title('Probability of 50.5% success rate with 50% accuracy')

# Sets grid for plot
ax.grid(linestyle='--', alpha=0.4)
# Ensures x-axis starts at 0
ax.set_xlim([0, None])

# Function used to format data to represent percentages
def to_percentage(integerVal, _):
  return str(integerVal)+'%'
# Set the y-axis data to end with a '%'
ax.yaxis.set_major_formatter(ticker.FuncFormatter(to_percentage))

# Displays plot
plt.show()
```
