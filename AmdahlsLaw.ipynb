```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# Function to calculate the speedup according to Amdahl's law
def amdahl(workers, parallelPortion):
    return 1 / (1 - parallelPortion + (parallelPortion/x))


# Set 10000 different work amounts between 1-750
x = np.linspace(1, 750, 10000)

# Set figure and axes objects for plot
fig, ax = plt.subplots()

# Plots data
for parallelPortion in [0.9, 0.95, 0.96, 0.97, 0.98, 0.99][::-1]:
  ax.plot(x, amdahl(x, parallelPortion), label=str(parallelPortion*100) + "% parallelisable")


# Plots the legend
ax.legend(loc='best')

# Plots a label for each axis
plt.ylabel('Speedup')
plt.xlabel('Workers')
# Plots the title
plt.title('Amdahl\'s law')

# Sets grid for plot
ax.grid(linestyle='--', alpha=0.4)
# Ensures x-axis starts at 0
ax.set_xlim([0, None])

fig.set_size_inches(10, 6)

# Displays plot
plt.show()
```
